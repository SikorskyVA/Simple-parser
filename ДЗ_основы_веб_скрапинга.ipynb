```python

import pandas as pd, requests, time
from bs4 import BeautifulSoup

query_list = ['python', 'анализ данных']
def get_articles_habr(query_):
    search_of_habr = pd.DataFrame()
    for query_ in query_list:
        URL = 'https://habr.com/ru/search/'
        params = {
            'q': query_
        }
        headers = {
            'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Mobile Safari/537.36'
        }
        req = requests.get(URL, params=params, headers=headers)
        soup = BeautifulSoup(req.text)
        articles = soup.find_all('article', class_='tm-articles-list__item')
        for article in articles:
            if article.find('div', class_='tm-megapost-snippet'):
                title = article.find('h2', class_='tm-megapost-snippet__title').text.strip()
                date = article.find('time', class_='tm-megapost-snippet__datetime-published').text.strip()
                link = 'https://habr.com' + article.find('a', class_='tm-megapost-snippet__link tm-megapost-snippet__card').get('href')
            else:
                title = article.find('h2', class_='tm-article-snippet__title tm-article-snippet__title_h2').text.strip()
                date = article.find('span', class_='tm-article-snippet__datetime-published').text
                link = 'https://habr.com' + article.find('a', class_='tm-article-snippet__title-link').get('href')
            row = {'дата': date, 'заголовок': title, 'ссылка на материал': link}
            search_of_habr = pd.concat([search_of_habr, pd.DataFrame([row])])
        time.sleep(1)
    return search_of_habr.drop_duplicates(subset='ссылка на материал').reset_index(drop=True)
get_articles_habr(query_list).head(50)
```
